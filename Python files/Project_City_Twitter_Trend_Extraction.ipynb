```python
# Trend data extraction from Twitter
# This class containes the code for extracting trend information related to cities through Twitter API
# Detailed description availble through inline comments

import tweepy
import twitter# pip install twitter
import json 
import pandas as pd
from urllib.parse import unquote
import time


# Storing relevant keys and token information in desired variables
consumerKey = "8kN7ITYYYK6bNkHKoViRPb7gJ"
consumerSecret = "g8IGSzKWckrYRzYjTgm8J2oNIywsmzcS8I9zW6WTrtIx28RGrE"
accessToken = "1088871960698908674-WsDWHHSB2jcaH9zMwaW5mNgpwIzpGM"
accessTokenSecret = "0pHfCkkkjEhWeQbm6tZ9eWHRLXzIh4dyNimAHXmuvYg6j"


auth = tweepy.OAuthHandler(consumerKey, consumerSecret)
auth.set_access_token(accessToken, accessTokenSecret)
# Initialising twitter api
api = tweepy.API(auth)



# Initialising below empty lists for storing player popularity related information
cities_id_list = []
tweet_volume_list = []
trend_topic_list = []


# Reading the city data with WOE_ID and storing it in a dataframe
cities_woe_input = pd.read_excel('WOE_ID.xlsx')
#Constructing a dictionary from two columns of the read data namely 'City_ID' and 'WOE_ID' adn stored in cities_woe_dict
cities_woe_dict= dict(zip(cities_woe_input.City_id, cities_woe_input.WOE_ID))


#Iterating through the items of dictionary 'cities_woe_dict'
for key, value in cities_woe_dict.items():
    # Try block used here in order to avoid exceptions because of lack of response for woe IDs of certain cities
    try:
        # Retrievingn the trend for the current woe ID associated with a city. For twitter rate limit pupose, the count has 
        # been restricted to 10
        results = api.trends_place(int(value), count = 20)
        # Iterating through the results
        for location in results:
            #Ierating through the trends withn the results
            for trend in location["trends"]:
                # Storing the city_id, trend topic and tweet volume in respective lists
                cities_id_list.append(key)
                trend_topic_list.append(trend["name"])
                tweet_volume_list.append( trend["tweet_volume"])
    # The exception will be handled for the city with no trend and next city will be evaluated 
    except:
        continue
        
        
# Creating a dataframe cities_trend_df to store trend information collected with regard to cities         
cities_trend_df = pd.DataFrame(data = cities_id_list, columns = ['City_ID'])
cities_trend_df['Tweet_Volume'] = tweet_volume_list
cities_trend_df['Trend_Topic'] = trend_topic_list
cities_trend_df['Trend_Topic_Modified'] = cities_trend_df['Trend_Topic'].str.lower()
cities_trend_df['Trend_Topic_Modified'] = cities_trend_df['Trend_Topic_Modified'].str.replace(' ', '')
cities_trend_df['Trend_Topic_Modified'] = cities_trend_df['Trend_Topic_Modified'].str.replace('#','')
cities_trend_df['Trend_ID'] = cities_trend_df['City_ID'].astype(str)+ "_"+ cities_trend_df['Trend_Topic_Modified'].astype(str)
del cities_trend_df['Trend_Topic_Modified']


# Exporting the data from  cities_trend_df dataframe into a csv by name 'Project_Twitter_Trend_Data.csv'
cities_trend_df.to_csv("Project_City_Twitter_Trend_Data.csv", index=False)



""""
CONCLUSIONS:

This code uses the list of cities along with woe IDs, selected for the project and  tries to find the trend associated with 
every city through Twitter API, stores the results in dataframe and exports it to a CSV file.

CONTRIBUTIONS:

RAJENDRA KUMAR RAJKUMAR - 75% 
MONISH  HIRISAVE RAGHU - 25%

CITATIONS:

1. https://www.geeksforgeeks.org
2. https://github.com/nikbearbrown/INFO_6210
3. http://woeid.rosselliot.co.nz/lookup/

The code with regard to extraction of information from twitter was used from the above mentioned resources

Original writtten code - 60%
Code referenced from external sources(but modified suiting needs) - 40%

LICENSE:

Copyright <2019> <RAJENDRA KUMAR RAJKUMAR, MONISH  HIRISAVE RAGHU>

Permission is hereby granted, free of charge, to any person obtaining a copy of this software and associated documentation files (the "Software"), to deal in the Software without restriction, including without limitation the rights to use, copy, modify, merge, publish, distribute, sublicense, and/or sell copies of the Software, and to permit persons to whom the Software is furnished to do so, subject to the following conditions:

The above copyright notice and this permission notice shall be included in all copies or substantial portions of the Software.

THE SOFTWARE IS PROVIDED "AS IS", WITHOUT WARRANTY OF ANY KIND, EXPRESS OR IMPLIED, INCLUDING BUT NOT LIMITED TO THE WARRANTIES OF MERCHANTABILITY, FITNESS FOR A PARTICULAR PURPOSE AND NONINFRINGEMENT. IN NO EVENT SHALL THE AUTHORS OR COPYRIGHT HOLDERS BE LIABLE FOR ANY CLAIM, DAMAGES OR OTHER LIABILITY, WHETHER IN AN ACTION OF CONTRACT, TORT OR OTHERWISE, ARISING FROM, OUT OF OR IN CONNECTION WITH THE SOFTWARE OR THE USE OR OTHER DEALINGS IN THE SOFTWARE.


"""
```

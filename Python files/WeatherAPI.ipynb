```python
import requests
import calendar

#Retrieving API response by using API call
api_key = '9cfe559b3545b304a5e7f920bdec0028'
api_call = 'https://api.openweathermap.org/data/2.5/forecast?appid=' + api_key

running = True

print('Welcome to Holiday Planner\'s 5 day weather forecast application using OpenWeatherMap\'s API!')
```

```python
import xlrd 

#Declaration of variables required for weather forecasting
city =[]
dates = []
tempr = []
tempr_min = []
tempr_max = []
cty = []
description = []
pressure = []
humidity = []
cloudiness = []
wind = []
forecast_id=[]
count=1


#Location of the excel file containg hotel links,city id and stateid
loc = ("C:/Users/prudh/Desktop/INFO 6210 DMDD/Project/hotels_urls.xlsx")

#opeing the workbook from the above mentioned location
wb = xlrd.open_workbook(loc) 

#Fetching data from sheet 1
sheet = wb.sheet_by_index(0)  


#Looping  over the cities in excel and appending it to a list
for i in  range(sheet.nrows):
    city.append(sheet.cell_value(i,1))
    
#Looping over the list of cities extracted from the excel sheet
for c in city:
    
    #Appending the city to the api call to fetch the response
    api_call += '&q=' + c
    json_data = requests.get(api_call).json()
    
    current_date = ''

    city_name=c
    # Iterates through the array of dictionaries named list in json_data
    for item in json_data['list']:
        
        #Autogeneration of forecast id
        forcstid = "wea"+str(count)
        forecast_id.append(forcstid)
        count=count+1

        # Time of the weather data received, partitioned into 3 hour blocks
        time = item['dt_txt']

        # Stores the current date and prints it once
        if current_date != time:
            current_date = time
        dates.append(current_date)
        
        #stores the currect temp in celsius
        temperature = item['main']['temp']
        t='{:.2f}'.format(temperature - 273.15)
        t=float(t)
        tempr.append(t)
        
        #Stores the city id
        cty.append(city_name)
        
        #Stores the min temperature
        tempmin = item['main']['temp_min']
        tmin = '{:.2f}'.format(tempmin - 273.15)
        tmin = float(tmin)
        tempr_min.append(tmin)
        
        #Stores the max temperature
        tempmax = item['main']['temp_max']
        tmax = '{:.2f}'.format(tempmax - 273.15)
        tmax = float(tmax)
        tempr_max.append(tmax)
        
        #Stores the current weather condition
        desc = item['weather'][0]['description']
        description.append(desc)
        
        #Stores the current pressure
        prs = item['main']['pressure']
        pressure.append(prs)
        
        #Stores the current humidity
        hum = item['main']['humidity']
        humidity.append(hum)
        
        #Stores the currect Clouiness percentage
        cld = item['clouds']['all']
        cloudiness.append(cld)
        
        #Stores the current wind speed in miles/hr
        wnd = item['wind']['speed']
        wnd = wnd*2.237
        wnd= float(wnd)
        wind.append(wnd)
```

```python
import pandas as ps

#Appending all the weather data into a data frame 
weather_data  = ps.DataFrame({  "Forecast Id":forecast_id,
                                "City id":cty,
                                "Date":dates,
                                "Temp(C)":tempr,
                                "Minimum Temperature(C)":tempr_min,
                                "Maximum Temperature(C)":tempr_max,
                                "Weather Condition":description,
                                "Pressure(hPa)": pressure,
                                "Humidity(%)":humidity,
                                "Cloudiness(%)":cloudiness,
                                "Wind(miles/hr)":wind})


#Display of weather data frame
weather_data
```

```python
export_csv = weather_data.to_csv(r'C:\Users\prudh\Desktop\INFO 6210 DMDD\Project\CSV files\Weather_API_latest.csv', index = None, header = True)
```

```python
import pyodbc
import pandas as ps
```

```python
conn1=pyodbc.connect("Driver={SQL Server Native Client 11.0};"
                    "server=LAPTOP-0QUGABL8;"
                    "database=holiday_planner;"
                   "Trusted_Connection=yes;")
```

```python
# Find attractions of a city based on city name and entry type input

curr = conn1.cursor()

city_name=input("Please enter the city which you want to visit:")
entry_type=input("Please enter you choice of attraction entry type-Free/Paid:")
params=(city_name,entry_type)
curr.execute("{CALL sp_attraction_city_type(?,?)}",params)


cityname=[]
attrname=[]
entrytype=[]
attrtype=[]
timespent=[]
attractionid=[]
state=[]
for row in curr:
    cityname.append(row.city_name)
    attrname.append(row.attraction_name)
    attrtype.append(row.attraction_type)
    entrytype.append(row.entry_type)
    timespent.append(row.Time_To_Spend)
    attractionid.append(row.attraction_id)
    state.append(row.state_id)
    
attraction_data  = ps.DataFrame({  "City Name":cityname,
                                 "Attraction Name":attrname,
                                "Entry_type":entrytype,
                                 "Attraction type":attrtype,
                                "Time you can spend":timespent,
                                "Attraction Id":attractionid,
                                "State":state}) 



curr.close()
conn1.close()
attraction_data
```

```python
"""
Use Case 2:
Display weather of a city for a particular date/time
"""
```

```python

conn2=pyodbc.connect("Driver={SQL Server Native Client 11.0};"
                    "server=LAPTOP-0QUGABL8;"
                    "database=holiday_planner;"
                   "Trusted_Connection=yes;")
curr = conn2.cursor()

city_name=input("Please enter the city which you want to forecast weather for:")
date=input("Please enter date:")
params=(city_name,date)
curr.execute("{CALL sp_weather_city_date(?,?)}",params)

cityname=[]
date=[]
temperature=[]
weather=[]
mint=[]
maxt=[]
pressure=[]
humidity=[]
cloud=[]
wind=[]

for row in curr:
    cityname.append(row.City_Name)
    date.append(row.Date)
    temperature.append(row.Temp_C_)
    weather.append(row.Weather_Condition)
    mint.append(row.Minimum_Temperature_C_)
    maxt.append(row.Maximum_Temperature_C_)
    pressure.append(row.Pressure_hPa_)
    humidity.append(row.Humidity___)
    cloud.append(row.Cloudiness___)
    wind.append(row.Wind_miles_hr_)

weather_data  = ps.DataFrame({  "City Name":cityname,
                                 "Date":date,
                                "Temperature(C)":temperature,
                                 "Weather Condition":weather,
                                "Minimum Temperature(C)":mint,
                                "Maximum Temperature(C):":maxt,
                                "Pressure(hPa)":pressure,
                                 "Humidity(%)":humidity,
                                 "Cloudiness":cloud,
                                 "Wind Speed(m/hr)":wind},) 



curr.close()
conn2.close()
weather_data
```

```python

"""
Use Case 4:
Display list of cheapest hotels for a city with approximate price/night.

"""
```

```python

conn4=pyodbc.connect("Driver={SQL Server Native Client 11.0};"
                    "server=LAPTOP-0QUGABL8;"
                    "database=holiday_planner;"
                   "Trusted_Connection=yes;")
curr = conn4.cursor()

city_name=input("Please enter the city which you want to search hotels for for:")

params=(city_name)
curr.execute("{CALL sp_hotels_cheapest(?)}",params)

cityname=[]
hotel=[]
rating=[]
reviewc=[]
approxprice=[]
bnum=[]
sname=[]
landmark=[]


for row in curr:
    cityname.append(row.city_name)
    hotel.append(row.hotel_name)
    if ((row.rating)=='no rating available'):
        rating.append('no rating available')
    else:
        rating.append(row.rating)
    reviewc.append(row.total_reviewers)
    if(row.price_per_night is None):
        approxprice.append('N/A')
    else:
        approxprice.append(row.price_per_night)
    bnum.append(row.building_number)
    sname.append(row.street_name)
    landmark.append(row.landmark)


hotel_data  = ps.DataFrame({  "City Name":cityname,
                                 "Hotel Name":hotel,
                                "Rating":rating,
                                "Total Reviewers":reviewc,
                                "Approximate price per night":approxprice,
                                "Building number":bnum,
                                "Street name":sname,
                                "Landmark":landmark}) 



curr.close()
conn4.close()
hotel_data
```

```python

"""
Use Case 5
Display flight schedules for origin, destination, month and day input

"""
```

```python

conn5=pyodbc.connect("Driver={SQL Server Native Client 11.0};"
                    "server=LAPTOP-0QUGABL8;"
                    "database=holiday_planner;"
                   "Trusted_Connection=yes;")
curr = conn5.cursor()

origin=input("Origin:")
destination=input("Destination:")
month=input("Month:")
day=input("Day:")

params=(origin,destination,month,day)
curr.execute("{CALL sp_flight_schedule(?,?,?,?)}",params)

Airline=[]
Origin=[]
Destination=[]
Departure=[]
Arrival=[]
Distance=[]
AirTime=[]



for row in curr:
    Airline.append(row.airline_name)
    Origin.append(row.Origin_City_ID)
    Destination.append(row.Destination_City_ID)
    if ((row.departure_time) is None):
        Departure.append('N/A')
    else:
        Departure.append(row.departure_time)
    if ((row.arrival_time) is None):
        Arrival.append('N/A')
    else:
        Arrival.append(row.arrival_time)
  
    if(row.distance is None):
        Distance.append('N/A')
    else:
        Distance.append(row.distance)
    if((row.Air_Time) is None):
        AirTime.append('N/A')
    else:
        AirTime.append(row.Air_Time)
        


flight_data  = ps.DataFrame({  "Airline":Airline,
                           "Origin":Origin,
                           "Destination":Destination,
                           "Departure Time":Departure,
                           "Arrival Time":Arrival,
                           "Distance":Distance,
                           "Air Time":AirTime}) 



curr.close()
conn5.close()
flight_data
```

```python

"""
Use Case 6
Display ratings of services offered by a particular airline

"""
```

```python
conn6=pyodbc.connect("Driver={SQL Server Native Client 11.0};"
                    "server=LAPTOP-0QUGABL8;"
                    "database=holiday_planner;"
                   "Trusted_Connection=yes;")
curr = conn6.cursor()

Airline=input(" Enter Airline Name:")


params=(Airline)
curr.execute("{CALL sp_airline_rating(?)}",params)

Airline=[]
food=[]
entertainment=[]
seating=[]
staff=[]
vmny=[]




for row in curr:
    Airline.append(row.airline_name)
    food.append(row.food_and_beverage_rating)
    entertainment.append(row.Inflight_Entertainment_Rating)
    seating.append(row.Seat_Comfort_Rating)
    staff.append(row.Staff_Service_Rating)
    vmny.append(row.Value_For_Money_Rating)
        


airline_data  = ps.DataFrame({  "Airline":Airline,
                               "Food":food,
                                "Entertainment":entertainment,
                                "Seat Comfort":seating,
                                "Staff Service":staff,
                                "Value for Money":vmny}) 



curr.close()
conn6.close()
airline_data
```
